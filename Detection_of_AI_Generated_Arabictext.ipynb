```python
#استيراد المكتبات
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

```python
dataset = load_dataset("KFUPM-JRCAI/arabic-generated-abstracts")
print(dataset)



dfs = []
for split in dataset.keys():
    temp_df = pd.DataFrame(dataset[split])

#إضافة عمود(source-split) لتوضيح من أين جاء النص بعد الدمج

human_df = pd.DataFrame({
    "text": temp_df["original_abstract"],
    "label": "human",
    "source_split": split
})


# إضافة عمود (generator)  لمعرفة أي نموذج أنتج النص.
ai_dfs = []
for col in ["allam_generated_abstract", "jais_generated_abstract", "llama_generated_abstract", "openai_generated_abstract"]:
    ai_df = pd.DataFrame({
        "text": temp_df[col],
        "label": "AI-generated",
        "source_split": split,
        "generator": col
    })
    ai_dfs.append(ai_df)


    #دمج النصوص
    dfs.append(pd.concat([human_df] + ai_dfs, ignore_index=True))
    final_df = pd.concat(dfs, ignore_index=True)


# عدد النصوص البشرية (label = human)
num_human = final_df[final_df["label"] == "human"].shape[0]

# عدد النصوص المولدة بالذكاء الاصطناعي (label = AI-generated)
num_ai = final_df[final_df["label"] == "AI-generated"].shape[0]

print("عدد النصوص البشرية:", num_human)
print("عدد النصوص المولدة بالذكاء الاصطناعي:", num_ai)




#   القيم الفارغة (Missing values)
print("عدد الخلايا الفارغة في كل عمود:")
print(final_df.isnull().sum())

#   التكرار (Duplicates)
num_duplicates = final_df.duplicated().sum()
print("\nعدد الصفوف المكررة:", num_duplicates)




```

```python
print("أول 5 نصوص بشرية:")
print(final_df[final_df["label"] == "human"][["text", "source_split"]].head(), "\n")



print(" أول 5 نصوص مولدة بالذكاء الاصطناعي:")
print(final_df[final_df["label"] == "AI-generated"][["text", "source_split", "generator"]].head())
```
